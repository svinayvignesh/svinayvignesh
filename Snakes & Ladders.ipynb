```python
# -*- coding: utf-8 -*-
"""
@author: S. Vinay Vignesh
1828185
Simple Snakes & Ladders Game
"""
#%%
import random
snakes = {17:7, 54:34, 62:19, 98:79}
ladders = {3:38, 24:33, 42:93, 72:84}
print("#### Welcome to Snakes & Ladders Game #### ")
print("########## by S. Vinay Vignesh ###########")
print("#"*42)
player1 = None
player2 = None
p1_current = 0
p2_current = 0
while not player1:
    player1 = input("Enter Player 1's name: ")
while not player2:
    player2 = input("Enter Player 2's name: ")
print("Welcome "+player1+" and "+player2)
print("#"*42)
print("Let's Begin:")
print("#"*42)
#%%
def ladder_boost(old, current, name):
    print( name + " climbed from " + str(old) + " to " + str(current))
def bitten(old, current, name):
    print( name + " got bitten down from " + str(old) + " to " + str(current))


#%%
def move_player(name, current, dice):
    old = current
    current = current + dice

    if current > 100:
        print(str(100 - old) + " steps needed.")
        return old

    print(name + " moved from " + str(old) + " to " + str(current))
    if current in snakes:
        final = snakes.get(current)
        bitten(current, final, name)

    elif current in ladders:
        final = ladders.get(current)
        ladder_boost(current, final, name)

    else:
        final = current

    return final 
#%%
def winner(name, position):
    if position == 100:
        print("Congrats, "+name+" has won the game.") 
        print("#"*42)
        print("\n Thankyou For Playing.\n")
        raise SystemExit("This is the only way to exit without importing sys")
#%%
def movement(player, current):
    value = input("\n"+ player + ": ")
    if type(value) == str and value == "roll":
        dice_number = random.randint(1,6)
    elif type(value) == str and value == "quit":
        print("Quitting\nThankyou for playing.")
        raise SystemExit
    elif value.isdigit() and int(value) > 0 and int(value) < 21:
        dice_number = int(value)
    else:
        print("Invalid response.")
        print("Enter 'roll' to roll dice or enter a value from 1 to 20.")
        return 0
    print(player + " got", dice_number)
    current = move_player(player, current, dice_number)
    print(player, " is at ", current)
    print("#"*42)
    winner(player, current)
    return current
    
while 1:
    temp = 0
    while temp == 0:
        temp = movement(player1, p1_current)
    p1_current = temp
    temp = 0
    while temp == 0:
        temp = movement(player2, p2_current)
    p2_current = temp
```
